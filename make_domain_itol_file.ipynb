```python
import Bio
from Bio import Seq
from Bio import SeqIO
```

```python
seqfile = 'allhits.fa'
accnfile = 'accn_list.txt'
domfile = 'domain_itol_colors.dmp'
lookupfile = 'new_accn_lookup.txt'
```

```python
fadict = {}
for seq_record in SeqIO.parse(seqfile, "fasta"):

    header = seq_record.description
    accn = header.split('-')[0]
    sequence = str(seq_record.seq)
    fadict[accn] = [header,sequence]
    
    #print(accn)
    
fi = open(accnfile)
accnDict = {}
for line in fi:
    accn = line.rstrip()
    accnDict[accn] = {}
    #print(accn)
    
fi.close()

fi = open(domfile)
domLookup = {}
for line in fi:
    col = line.rstrip().split('\t')
    domain = col[0]
    color = col[2]
    shape = col[4]
    
    domLookup[domain] = [color,shape]
    
    print(domain,color,shape)
    
fi.close()

fi = open(lookupfile)
lookup = {}
for line in fi:
    header = line.rstrip()
    accn = header.split('-')[0]
    
    if accn not in lookup:
        lookup[accn] = set()
    lookup[accn].add(header)
    
    #print(accn,header,fadict[accn][0])
fi.close()

```

```python
hmmfile = 'ALL_IUPRED_CUSTOM.domtbl.out'
fi = open(hmmfile)

bigDict = {}

for line in fi:
    if line[0] == '#':
        continue
    col = line.rstrip().split()
    domain = col[3].split('_')[0].split('.')[0]

    if domain not in domLookup:
        print(domain)
        continue
    
    seq = col[0]
    accn = seq.split('-')[0]
    if accn not in accnDict:
        continue
        
    evalue = float(col[11])
    if evalue > 1e-3:
        continue
        
    start = int(col[17]) - 1
    end = int(col[18])
    
    if domain not in accnDict[accn]:
        accnDict[accn][domain]= {}
    accnDict[accn][domain][start] = [evalue, start, end]

fi.close()
```

```python
maxsnf = 0
for accn in accnDict:
    if 'SNF2' in accnDict[accn]:
        for domain in accnDict[accn]['SNF2']:
            end = accnDict[accn]['SNF2'][domain][2]
            if end > maxsnf:
                maxsnf = end

print(maxsnf)
```

```python
minlen = 50
for accn in accnDict:
        
    spacerlen = 0    
    if 'SNF2' not in accnDict[accn]:
        # median length of SNF2 is 316 
        spacerlen = maxsnf - accnDict[accn]['Helicase'][min(accnDict[accn]['Helicase'])][1] + 10
        #print(spacerlen)
        
    else:
        spacerlen = maxsnf - accnDict[accn]['SNF2'][max(accnDict[accn]['SNF2'])][2]
        #print(spacerlen)
        
    #header = fadict[accn][0]
    length = str(len(fadict[accn][1]) + spacerlen)
    #fo.write(header + ',' + length)
    #print(accn,length)

    domdesc = ''
    if spacerlen > 0:
        domdesc = domdesc + ',' + 'RE' + '|' + '0' + '|' + str(spacerlen) + '|' + '#FFFFFF' + '|' + 'spacer'
        
    for domtype in accnDict[accn]:
        for domain in accnDict[accn][domtype]:
            start = str(accnDict[accn][domtype][domain][1] + spacerlen)
            end = str(accnDict[accn][domtype][domain][2] + spacerlen)

            if accnDict[accn][domtype][domain][2] - accnDict[accn][domtype][domain][1] < minlen:
                end = str(accnDict[accn][domtype][domain][1] + spacerlen + minlen)

            color = domLookup[domtype][0]
            shape = domLookup[domtype][1]
            domdesc = domdesc + ',' + shape + '|' + start + '|' + end + '|' + color + '|' + domtype
        #print(accn,domain,start,end,header,color,shape)
        
    #fo.write('\n')
    #print(accn, length, domdesc)
    if accn in lookup:
        for header in lookup[accn]:
            #print(header, length, domdesc)
            fo.write(header + ',' + length + domdesc + '\n')
    
fo.close()
```
